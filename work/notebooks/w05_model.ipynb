## Train and Compare Against the Baseline

The Random Forest model was trained using the selected historical search performance features and evaluated on the same test split as the Week 4 baseline rule.

```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Load dataset
df = pd.read_csv("/content/flyrank_assign_1/data/raw/content_refresh_anonymized.csv")

# Create target
df["is_declining"] = (df["trend_direction"] == "down").astype(int)

# Features
feature_cols = [
    "impressions_90d",
    "clicks_90d",
    "avg_position",
    "ctr",
    "content_age_days",
    "search_volume",
    "engagement_rate"
]

X = df[feature_cols]
y = df["is_declining"]

# Train/Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

# Train model
model = RandomForestClassifier(
    n_estimators=200,
    random_state=42
)

model.fit(X_train, y_train)

# Predict
pred = model.predict(X_test)

# Metrics
print("Accuracy :", accuracy_score(y_test, pred))
print("Precision:", precision_score(y_test, pred))
print("Recall   :", recall_score(y_test, pred))
print("F1 Score :", f1_score(y_test, pred))
```

```python
comparison = pd.DataFrame({
    "Model": ["Baseline Rule", "Random Forest"],
    "Accuracy": [
        0.6763333333333333,
        accuracy_score(y_test, pred)
    ],
    "Precision": [
        0.6897450753186558,
        precision_score(y_test, pred)
    ],
    "Recall": [
        0.7321648216482165,
        recall_score(y_test, pred)
    ],
    "F1 Score": [
        0.7103221957040573,
        f1_score(y_test, pred)
    ]
})

comparison
```

## Errors and Interpretation

The Random Forest correctly identifies many pages that are likely to require review, but it still produces both false positives and false negatives.

False positives cause healthy pages to be unnecessarily reviewed, increasing analyst workload.

False negatives are more costly because pages experiencing genuine performance decline may not be reviewed in time.

Feature importance suggests that historical impressions, clicks, average position, and CTR contribute most strongly to the model's predictions. Compared with the Week 4 baseline, the model captures more complex interactions between these signals and therefore produces better overall ranking performance.
